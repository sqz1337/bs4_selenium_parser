```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
browser = webdriver.Chrome('C:\webdr\chromedriver.exe')

browser.get('https://legalbet.ru/complaints/bk-leon/')
```

```python
for i in range(3):
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    share =  browser.find_element(By.XPATH, '//*[@id="infinite-list"]/a/div')
    share.click()
    time.sleep(3)
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```

```python
source_data = browser.page_source
soup = bs(source_data)
```

```python
review = soup.find_all('div', class_="description")
```

```python
review_text_child = []
for i in range(3000):
    children = review[i].findChildren("a" , recursive=False)
    for child in children:
        review_text_child.append(child)
```

```python
review_text_child
```

```python
review_text_child_clean_a = []
paragraph = [a.text for a in review_text_child]
review_text_child_clean_a.append(paragraph)
```
